# MULTITHREADING

## When to use?

 I/O bound tasks : Tasks that spend more time waiting for I/O operation(e.g. file operation) 

 Concurrent Execution : When you want to improve the throughput of your application by processing

```python

import threading
import time

def print_numbers():
    for i in range(5):
        time.sleep(2)
        print(f"Number: {i}")

def print_letters():
    for letter in "abcde":
        time.sleep(2)
        print(f"Letter: {letter}")


t = time.time()

print_numbers()
print_letters()
finished_time = time.time() - t
print(finished_time)

```

```python
## I want execution for print_letters() when print_numbers() is in sleep

## create two thread


t1 = threading.Thread(target=print_numbers)
t2 = threading.Thread(target=print_letters)

t = time.time()
## Lets strat the thread
t1.start()
t2.start()

## Wait for thread to complete
t1.join()
t2.join()
finished_time = time.time() - t
print(finished_time)
```

```python
## FINAL CODE IN BETTER WAY

import threading
import time

print_lock = threading.Lock()

def print_numbers():
    for i in range(5):
        time.sleep(2)
        with print_lock:
            print(f"Number: {i}")

def print_letters():
    for letter in "abcde":
        time.sleep(2)
        with print_lock:
            print(f"Letter: {letter}")

t1 = threading.Thread(target=print_numbers)
t2 = threading.Thread(target=print_letters)

t = time.time()
## Lets strat the thread
t1.start()
t2.start()

## Wait for thread to complete
t1.join()
t2.join()
finished_time = time.time() - t
print(finished_time)
```
