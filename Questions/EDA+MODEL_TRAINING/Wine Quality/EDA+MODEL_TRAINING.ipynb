```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('wine_purity_dataset(in).csv')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df['alcohol'] = df['alcohol'].fillna(df['alcohol'].mean())
df['sulphates'] = df['sulphates'].fillna(df['sulphates'].mean())
df['chlorides'] = df['chlorides'].fillna(df['chlorides'].mean())
df['density'] = df['density'].fillna(df['density'].mean())
df['citric_acid'] = df['citric_acid'].fillna(df['citric_acid'].mean())
df['fixed_acidity'] = df['fixed_acidity'].fillna(df['fixed_acidity'].mean())
df['free_sulfur_dioxide'] = df['free_sulfur_dioxide'].fillna(df['free_sulfur_dioxide'].mean())
df['total_sulfur_dioxide'] = df['total_sulfur_dioxide'].fillna(df['total_sulfur_dioxide'].mean())
df['temperature'] = df['temperature'].fillna(df['temperature'].mean())

df['quality_grade'] = df['quality_grade'].fillna(df['quality_grade'].mode()[0])
df['aging_level'] = df['aging_level'].fillna(df['aging_level'].mode()[0])
df['yeast_strain'] = df['yeast_strain'].fillna(df['yeast_strain'].mode()[0])
```

```python
df.isna().sum()
```

```python
df=df.drop('region',axis=1)
```

```python
df.head()
```

```python
X=df.drop('purity',axis=1)
y=df['purity']
```

```python
X=pd.DataFrame(X)
X.head()
```

```python
num_features = [feature for feature in X.columns if df[feature].dtype!='O']
categorical_features = [feature for feature in X.columns if df[feature].dtype=='O']

ordinal_features = ['quality_grade','aging_level','yeast_strain']

nominal_features = [col for col in categorical_features if col not in ordinal_features]

print(num_features)
print(categorical_features)

print(ordinal_features)
print(nominal_features)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler

ohe=OneHotEncoder()
oe=OrdinalEncoder()

preprocessor = ColumnTransformer([
                ('ohe',ohe,nominal_features),
                ('oe',oe,ordinal_features)
                ],remainder='passthrough')
```

```python
X=preprocessor.fit_transform(X)
```

```python
X=pd.DataFrame(X,columns=preprocessor.get_feature_names_out())
```

```python
X.head()
```

```python
scaler = StandardScaler()
X=scaler.fit_transform(X)
```

```python
X=pd.DataFrame(X)
```

```python
X.head()
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=22)
X_pca=pca.fit_transform(X)

cum_var = np.cumsum(pca.explained_variance_ratio_)
n_components = np.argmax(cum_var>=0.95) + 1
print(n_components)
```

```python
X_pca = pd.DataFrame(X_pca)
X_pca.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_pca,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
lr=LinearRegression()

lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
print(r2_score(y_test,y_pred))
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures(degree=2,include_bias=True)
X_train_poly=poly.fit_transform(X_train)
X_test_poly=poly.transform(X_test)
```

```python
lr1=LinearRegression()

lr1.fit(X_train_poly,y_train)
y_pred1=lr1.predict(X_test_poly)
print(r2_score(y_test,y_pred1))
```
