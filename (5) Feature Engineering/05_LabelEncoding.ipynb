### Label Encoding
Label encoding is a fundamental data preprocessing technique used to convert categorical data into a numerical format suitable for machine learning models. Many algorithms cannot process non-numeric values, making encoding a necessary step when working with features such as colors, cities or product types.

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
## Create a simple data frame
df = pd.DataFrame({
    'color':['red','blue','green','green','red','blue']
})
```

```python
df.head()
```

```python
lbl_encoder = LabelEncoder()
```

```python
lbl_encoder.fit_transform(df[['color']])
```

```python
lbl_encoder.transform([['blue']])
```

```python
lbl_encoder.transform([['red']])
```

### Ordinal Encoding

Ordinal encoding is a technique used to convert categorical data into numerical values while preserving the inherent order or hierarchy among the categories. This is particularly useful in machine learning, as many algorithms perform better with numerical data. For example, in a dataset with shirt sizes like "Small," "Medium," and "Large," ordinal encoding can assign values such as 1, 2, and 3, respectively, to reflect their natural ranking.

```python
import pandas as pd
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df1=pd.DataFrame({
    'size':['small','medium','large','medium','small','large']
})
```

```python
df1
```

```python
encoder = OrdinalEncoder(categories=[['small','medium','large']])
```

```python
encoder.fit_transform(df1[['size']])
```

```python
encoder.transform([['small']])
```
