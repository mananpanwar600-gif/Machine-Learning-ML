### SCRATCH

```python
class RRegressor:
    def __init__(self, alpha):
        self.alpha = alpha
        self.intercept_=None
        self.coef_=None

    def fit(self, X_train, y_train):
        X = np.hstack((np.ones((X_train.shape[0], 1)), X_train))
        print(X)

        I = np.eye(X.shape[1])
        I[0, 0] = 0   

        m = np.linalg.inv(X.T @ X + self.alpha * I) @ X.T @ y_train

        self.intercept_ = m[0]
        self.coef_ = m[1:]
```

```python
ridge1=RRegressor(10)
ridge1.fit(X_train,y_train)
print(ridge1.coef_)
print(ridge1.intercept_)
```

### Gradient Descent

```python
import numpy as np

class GDregressor:
    def __init__(self, eta=0.01, epoch=1000):
        self.eta = eta
        self.epoch = epoch
        self.w = None
        self.b = None

    def fit(self, X_train, y_train):
        m, n = X_train.shape

        X = np.hstack((np.ones((m, 1)), X_train))

        self.w = np.zeros(n + 1)

        for i in range(self.epoch):
            y_pred = X @ self.w
            error = y_pred - y_train

            dw = (2/m) * (X.T @ error)

            self.w -= self.eta * dw

        self.b = self.w[0]
        self.w = self.w[1:]
```

```python
ridge2=GDregressor()
ridge2.fit(X_train,y_train)
print(ridge2.w)
print(ridge2.b)
```
