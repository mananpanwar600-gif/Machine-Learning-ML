## TAKING A NEW CSV FILE


```python
d_frame = pd.read_csv("Sales.csv")
d_frame
```

```python
d_frame.describe()
```

```python
d_frame.isnull().sum()
```

## Filling Missing values with the mean of the column

```python
d_frame['Retail_PricefillNA'] = d_frame['Retail Price (USD)'].fillna(d_frame['Retail Price (USD)'].mean())
```

```python
d_frame.head()
```

```python
d_frame.dtypes
```

## Rename Columns


```python
d_frame = d_frame.rename(columns={'Date':'Ship Date'})
d_frame.head()
```

## Change Data Type


```python
d_frame['Tax_new'] = d_frame['Tax (USD)'].astype(int)
d_frame.head()
```

```python
d_frame['New Price'] = d_frame['Retail Price (USD)'].apply(lambda x:x*2)
d_frame.head()
```

## Data Aggregating And Grouping

```python
d_frame.head()
```

```python
group_mean = d_frame.groupby('Customer Name')['Retail Price (USD)'].mean()
print(group_mean)
```

```python
group_sum = d_frame.groupby(['Order No','Order Quantity'])['Total (USD)'].sum()
print(group_sum)
```

## Aggregate mupltiple functions

```python
grouped_agg = d_frame.groupby('Order Quantity')['Total (USD)'].agg(['mean','sum','count'])

grouped_agg
```

## Merging And Joining Data Frames

```python
# Create some sample data frames

df1 = pd.DataFrame({'Key':['A','B','C'] , 'Value1' : [1,2,3]})
df2 = pd.DataFrame({'Key':['A','B','D'] , 'Value2' : [4,5,6]})

df1
```

```python
df2
```

```python
## Merge This data frame on the 'Key Coulums'

pd.merge(df1,df2,on="Key",how="inner")
```

```python
pd.merge(df1,df2,on="Key",how="outer")
```

```python
pd.merge(df1,df2,on="Key",how="left")
```

```python
pd.merge(df1,df2,on="Key",how="right")
```
