## PYTHON LOGGING
Python's logging module is a powerful and flexible way to track events in your application. It allows you to record messages at different severity levels, making debugging and monitoring easier

```python
import logging

## Configure basic logging settings
logging.basicConfig(level=logging.DEBUG)

## log messages with different severity levels
logging.debug("This is a debug message")
logging.info("This is info message")
logging.warning("This is warning message")
logging.error("This is an error message")
logging.critical("This is a critical message")
```

### Log Levels

1. DEBUG (Level: 10)
Purpose: Detailed information, typically of interest only when diagnosing problems.
Example Use: Debugging code, tracking variable values, or understanding program flow.
Message Example: "DEBUG: Variable x has value 42"

2. INFO (Level: 20)
Purpose: Confirmation that things are working as expected.
Example Use: General application flow or high-level events like starting/stopping services.
Message Example: "INFO: Application started successfully"

3. WARNING (Level: 30)
Purpose: An indication that something unexpected happened, or indicative of some problem in the near future (e.g., ‘disk space low’). The software is still working as expected.
Example Use: Non-critical issues that should be addressed.
Message Example: "WARNING: Disk space is below 10%"

4. ERROR (Level: 40)
Purpose: A more serious problem, the software has not been able to perform some function.
Example Use: Logging exceptions or failed operations.
Message Example: "ERROR: Failed to connect to database"

5. CRITICAL (Level: 50)
Purpose: A very serious error, indicating that the program itself may be unable to continue running.
Example Use: Logging system crashes or unrecoverable errors.
Message Example: "CRITICAL: System is out of memory"

```python
## Configuring the logging
import logging

logging.basicConfig(
    filename='app.log',
    filemode='w',
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)


logging.debug("This is a debug message")
logging.info("This is info message")
logging.warning("This is warning message")
logging.error("This is an error message")
logging.critical("This is a critical message")
```
