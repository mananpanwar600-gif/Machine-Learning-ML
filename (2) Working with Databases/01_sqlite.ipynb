### SQL
1. SQL is Structured Query Language used to query Relational Database System. It is written in C language.
2. SQL is standard which specifies how relational schema is created, data is inserted or updated in relations, transactions are started and stopped, etc.
3. Main components of SQL are Data Definition Language(DDL), Data Manipulation Language(DML), Data Control Language(DCL).

### SQLite
1. SQLite is an Relational Database Management System which is written in ANSI-C.
2. SQLite is file-based. It is different from other SQL databases because unlike most other SQL databases, SQLite does not have separate server process.
3. SQLite supports many features of SQL and has high performance but does not support stored procedures.

```python
import sqlite3
```

```python
## Connect to an SQLite database
connection=sqlite3.connect('example.db')
connection
```

```python
cursor = connection.cursor()
```

```python
## Create a table
cursor.execute('''
Create Table If Not Exists employees(
    id Integer Primary Key,
    name Text Not Null,
    age Integer,
    department text
    )
''')

## commit the changes
connection.commit()
```

```python
cursor.execute('''
Select * from employees
    
''')
```

```python
## Insert the data in sqlite table
cursor.execute('''
Insert Into employees(name,age,department)
               values('John','32','Data Scientist')
''')

cursor.execute('''
Insert Into employees(name, age, department)
               values('Alice', '28', 'Software Engineer')
''')

cursor.execute('''
Insert Into employees(name, age, department)
               values('Michael', '40', 'Product Manager')
''')

## Commit the changes
connection.commit()
```

```python
## Query the data from the table

cursor.execute('Select * from employees')
rows = cursor.fetchall()

## print queried data

for row in rows:
    print(row)
```
