```python
import numpy as np
import pandas as pd

np.random.seed(42)
N = 113000

# ---------------------- BASIC FEATURES ----------------------
age = np.random.randint(18, 70, N)

income = np.random.normal(50000, 20000, N)
income = np.clip(income, 5000, 200000)

gender = np.random.choice(['Male', 'Female', 'Other'], N, p=[0.48, 0.48, 0.04])

city_tier = np.random.choice(['Tier1', 'Tier2', 'Tier3'], N, p=[0.3, 0.4, 0.3])

education_level = np.random.choice(
    ['High School', 'Bachelors', 'Masters', 'PhD'], N
)

marital_status = np.random.choice(['Single', 'Married', 'Divorced'], N)

occupation = np.random.choice(
    ['Student', 'Engineer', 'Manager', 'Sales', 'Unemployed'], N
)

# ---------------------- USAGE FEATURES ----------------------
monthly_usage_hours = np.random.normal(120, 50, N)
monthly_usage_hours = np.clip(monthly_usage_hours, 5, 300)

num_logins = np.random.poisson(30, N)

avg_session_time = np.random.normal(20, 10, N)
avg_session_time = np.clip(avg_session_time, 1, 120)

# ---------------------- SUBSCRIPTION ----------------------
subscription_fee = np.random.choice([199, 299, 399, 499, 699], N)

payment_method = np.random.choice(
    ['Credit Card', 'Debit Card', 'UPI', 'Net Banking'], N
)

device_type = np.random.choice(['Mobile', 'Desktop', 'Tablet'], N)

internet_type = np.random.choice(['4G', '5G', 'WiFi'], N)

region = np.random.choice(['North', 'South', 'East', 'West'], N)

# ---------------------- CUSTOMER HISTORY ----------------------
customer_tenure = np.random.randint(1, 60, N)

num_complaints = np.random.poisson(2, N)

last_login_days = np.random.randint(0, 30, N)

discount_used = np.random.choice([0, 1], N)

referral = np.random.choice([0, 1], N)

# ---------------------- APP FEATURES ----------------------
app_version = np.random.choice(['v1', 'v2', 'v3', 'v4'], N)

preferred_content = np.random.choice(
    ['Movies', 'Sports', 'News', 'Music'], N
)

peak_usage_time = np.random.choice(
    ['Morning', 'Afternoon', 'Evening', 'Night'], N
)

customer_support_calls = np.random.poisson(1, N)

# ---------------------- CHURN LOGIC ----------------------
churn_prob = (
    0.25 * (monthly_usage_hours < 50) +
    0.2 * (num_logins < 10) +
    0.2 * (income < 30000) +
    0.1 * (city_tier == 'Tier3') +
    0.15 * (num_complaints > 3) +
    0.1 * (last_login_days > 15)
)

churn_prob = np.clip(churn_prob, 0, 0.95)
churn = np.random.binomial(1, churn_prob)

# ---------------------- OUTLIERS ----------------------
outlier_idx1 = np.random.choice(N, int(0.02 * N), replace=False)
income[outlier_idx1] *= 5

outlier_idx2 = np.random.choice(N, int(0.01 * N), replace=False)
monthly_usage_hours[outlier_idx2] *= 3

# ---------------------- MISSING VALUES FUNCTION ----------------------
def insert_missing(arr, p):
    arr = arr.astype('object')
    mask = np.random.rand(len(arr)) < p
    arr[mask] = np.nan
    return arr

income = insert_missing(income, 0.1)
avg_session_time = insert_missing(avg_session_time, 0.08)
gender = insert_missing(gender, 0.05)
education_level = insert_missing(education_level, 0.07)

# ---------------------- ENGINEERED FEATURES ----------------------
engagement_score = (
    0.4 * (monthly_usage_hours / 300) +
    0.3 * (num_logins / 100) +
    0.2 * (customer_tenure / 60) +
    np.random.normal(0, 0.05, N)
)

risk_score = (
    0.3 * (num_complaints / 10) +
    0.3 * (last_login_days / 30) +
    0.2 * (1 - discount_used) +
    np.random.normal(0, 0.05, N)
)

lifetime_value = income * customer_tenure * np.random.uniform(0.5, 1.5, N)

# ---------------------- DATAFRAME ----------------------
df = pd.DataFrame({
    'age': age,
    'income': income,
    'gender': gender,
    'city_tier': city_tier,
    'education_level': education_level,
    'marital_status': marital_status,
    'occupation': occupation,
    'monthly_usage_hours': monthly_usage_hours,
    'num_logins': num_logins,
    'avg_session_time': avg_session_time,
    'subscription_fee': subscription_fee,
    'payment_method': payment_method,
    'device_type': device_type,
    'internet_type': internet_type,
    'region': region,
    'customer_tenure': customer_tenure,
    'num_complaints': num_complaints,
    'last_login_days': last_login_days,
    'discount_used': discount_used,
    'referral': referral,
    'app_version': app_version,
    'preferred_content': preferred_content,
    'peak_usage_time': peak_usage_time,
    'customer_support_calls': customer_support_calls,
    'engagement_score': engagement_score,
    'risk_score': risk_score,
    'lifetime_value': lifetime_value,
    'churn': churn
})

# ---------------------- SHUFFLE ----------------------
df = df.sample(frac=1).reset_index(drop=True)

# ---------------------- SAVE ----------------------
df.to_csv("data.csv", index=False)

# ---------------------- OUTPUT ----------------------
print(df.shape)
print(df.head())
```
