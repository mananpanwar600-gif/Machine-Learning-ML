## Basic Indexing and Slicing
### 1D array


```python
arr = np.arange(1,10)
arr
```

```python
print(arr[5])
print(arr[5:8])
```

```python
arr[5:8] = 12
arr
```

```python
arr_slice = arr[5:8]
arr_slice
```

```python
arr_slice[1] = 12345
arr
```

```python
arr_slice[:] = 64
arr
```
