```python
X=df.drop('total_bill',axis=1)
y=df['total_bill']
```

```python
le = LabelEncoder()
cols = ['sex','smoker','time']

for col in cols:
    X[col] = le.fit_transform(X[col])
```

```python
ohe=OneHotEncoder()
encoded = ohe.fit_transform(X[['day']]).toarray()
```

```python
encoded_df = pd.DataFrame(encoded,columns=ohe.get_feature_names_out())
```
