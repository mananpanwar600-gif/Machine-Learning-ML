```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Algerian_fire.csv',header=1)
```

### Data Cleaning

```python
df[df.isna().any(axis=1)]
```

Data set is divided in two sets based on region from 0 to 121 we have Bejaia and from 122 we have Sidi-Bel abbes so we can make a new column based on region anbd assgin with labels like 0-> Bejaia,
1-> Abbes and can name that column as region

```python
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
```

```python
df.info()
```

```python
df[['Region']]=df[["Region"]].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
df.tail()
```

```python
df.isna().sum()
```

```python
df[df.isna().any(axis=1)]
```

```python
df=df.dropna().reset_index(drop=True)
```

```python
df.isna().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## Fix spaces in column name
df.columns=df.columns.str.strip()
```

```python
df.columns
```

```python
df.head()
```

```python
## Change the req columns as integer datatype
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
objects=[features for features in df.columns if df[features].dtype=='O']
```

```python
for i in objects:
    if i!='Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
## Lets save the cleaned dataset
df.to_csv('Algerian_fire_cleaned.csv',index=False)
```

### EDA

```python
# drop day month year
df_copy=df.drop(columns=['day','month','year'],axis=1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
## Here there are also spaces in like fire    ,not fire    so they are taken as different classes
```

```python
## encoding of categories of classes
df_copy['Classes']=np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
## Density plot
df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
## percentage for pie chart
plt.figure(figsize=(12,7))
plt.pie(df_copy['Classes'].value_counts(normalize=True)*100,labels=['Fire','Not Fire'],autopct='%1.1f%%')
plt.title('Classes')
plt.show()
```

```python
### Correlation
df_copy.corr()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df_copy.corr(),annot=True)
```

```python
sns.boxplot(df_copy['FWI'],)
```

```python
### Monthly fire analysis
df['Classes']=np.where(df['Classes'].str.contains('not fire'),'Not Fire','Fire')
dftemp=df.loc[df['Region']]==0
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=df)
plt.ylabel('Number of fires',weight='bold')
plt.xlabel('Months',weight='bold')
plt.title('Bejaia',weight='bold')
```

```python
dftemp=df.loc[df['Region']]==1
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=df)
plt.ylabel('Number of fires',weight='bold')
plt.xlabel('Months',weight='bold')
plt.title('Abbes',weight='bold')
```
